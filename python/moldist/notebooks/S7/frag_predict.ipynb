```python
from src.features.build_features import buildFragment

# Build the features
Xfull,fpcat = buildFragment([t[1] for t in targets])
```

```python
# Build the targets
# yfull[i][j] is 1 if fold change is definitely >2, -1 if it is definitely less, and 0 if indeterminate
# i - compound, j - aptamer
yfull=[[0 if i[1]<2 else 1 if i[0]>2 else np.nan for i in zip(ff[0],ff[1])] for ff in fold]
# Transpose
yfull=np.array(yfull).T
```

```python
print(yfull.shape)
```

```python
# Choose an aptamer to work on
# Setup training data to include only the ones for which we have definite hit/miss classification
apt=1
tsel=[True if np.isfinite(yi) else False for yi in yfull[apt]]
y=yfull[apt][tsel]
X=[Xfull[i] for i in range(len(tsel)) if tsel[i]]
print(f"Aptamer {aptamers[apt]} has {sum(y):.0f} positives")
```

```python
tnames=[f"{targets[i][0]}[{fold[i][0][apt]:.2f},{fold[i][1][apt]:.2f}]" for i in range(len(tsel)) if tsel[i] ]
#print(tnames)
```

```python
print(f"Testing {aptamers[apt]} against {len(tnames)} targets with {sum(y):.0f} positives")
```

```python
from src.models import split_data

# Create train/test sets
seed=2
X_train, X_test, y_train, y_test, ind_train, ind_test = split_data.split(0.4,X,y,seed)
print(f"Have {sum(y_train):.0f} hits in train, {sum(y_test):.0f} in test.")
tname_test=[tnames[x] for x in ind_test]
tname_train=[tnames[x] for x in ind_train]
```

```python
from src.models import train_model

# Train model
# Need to handle NaNs!
models=[train_model.train_rfc(X_train,y_train,seed,n_estimators=10),train_model.train_rfc(X_train,y_train,seed,n_estimators=100)]
```

```python
from src.models import predict_model

# Test model
for model in models:
    print("------")
    print(model)
    print("Train:")
    yp_train=predict_model.predict(model, X_train, y_train, tname_train)
    print("Test:")
    yp_test=predict_model.predict(model, X_test, y_test, tname_test)
    print("LOO:")
    yp_loo=predict_model.predictLOO(model,X,y,tnames)
```

```python
from src.models import bitranking

# Rank fragments based on observed activity
ntop=20
for apt in range(len(aptamers)):
    print("Testing",aptamers[apt])
    tsel=[True if np.isfinite(yi) else False for yi in yfull[apt]]
    y=yfull[apt][tsel]
    X=[Xfull[i] for i in range(len(tsel)) if tsel[i]]
    tn=[targets[i][0] for i in range(len(tsel)) if tsel[i] ]

    frag=bitranking.getFragRanks(fpcat, X, y,ntop=ntop)
    print('Postives       : ',[tn[i] for i in range(len(tn)) if y[i]==1 and X[i].GetBit(int(frag))])
    print('False negatives: ',[tn[i] for i in range(len(tn)) if y[i]==1 and not X[i].GetBit(int(frag))])
    print('False positives: ',[tn[i] for i in range(len(tn)) if y[i]==0 and X[i].GetBit(int(frag))])
    print()
```
